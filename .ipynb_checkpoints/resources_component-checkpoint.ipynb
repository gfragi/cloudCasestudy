```python
# %%============== Import libraries =========
import numpy as np
import pandas as pd
import warnings
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import stats
from scipy import stats
import plotly.express as px
import my_functions as mf
import plotly.io as pio
import random
import vs_functions as vl

pio.renderers.default = "browser"
import plotly.graph_objects as go
```

```python
# %%==============  Define the initial values  =========

range1 = random.sample(range(10, 1024), 100)
range2 = random.sample(range(1025, 5000), 100)
range3 = random.sample(range(5001, 10000), 100)
range4 = random.sample(range(10001, 80000), 100)

tot_users  = range1 + range2 + range3 + range4
tot_users
iaas_dict = dict()
caas_dict = dict()
```

```python
disk_rand = list()

disk_list = ['S', 'H']

for i in range(1,401):
    disk_rand += random.choice(disk_list)
    

disk_type = pd.DataFrame(disk_rand)
disk_type = disk_type.replace({'S': 'SSD', 'H': 'HDD'})
```

```python
# %%==============  Calculate the total resources per technology by users   =========
vm_types = ['xsmall', 'small', 'medium', 'large'] 
components = ['nginx', 'framework', 'rel_db']

# users_size = zip(users, vm_types)
# users_size
```

```python
for users in tot_users:
    iaas_dict[users, 'nginx'] = vl.iaas.nginx(users)
    iaas_dict[users, 'framework'] = vl.iaas.framework(users)
    iaas_dict[users, 'rel_db'] = vl.iaas.rel_db(users)
    caas_dict[users, 'nginx'] = vl.caas.nginx(users)
    caas_dict[users, 'framework'] = vl.caas.framework(users)
    caas_dict[users, 'rel_db'] = vl.caas.rel_db(users)
    caas_dict[users, 'mem_db'] = vl.caas.memory_db(users)
    
# print("iaas_dict", iaas_dict)
# print("caas_dict", caas_dict)
```

```python
iaas_tot = dict()
caas_tot = dict()

for users in tot_users:

    iaas_tot[users] = vl.iaas.total_inst(users)
    caas_tot[users] = vl.caas.total_inst(users)
    
    
# print (iaas_tot)
```

```python
# Convert dictionary to dataframe
iaas = pd.DataFrame.from_dict(iaas_dict)
caas = pd.DataFrame.from_dict(caas_dict)
iaas_inst = pd.DataFrame(iaas_tot)
caas_inst = pd.DataFrame(caas_tot)
```

```python
# Transpose the dataframes
iaas = iaas.transpose()
caas = caas.transpose()
iaas_total = iaas_inst.transpose()
caas_total = caas_inst.transpose()
```

```python
# Make index to column
iaas.reset_index(inplace=True)
caas.reset_index(inplace=True)
iaas_total.reset_index(inplace=True)
caas_total.reset_index(inplace=True)
iaas
```

```python
iaas = iaas.replace({ 'framework': 'NodeJs', 'rel_db': 'postgres'})
iaas
```

```python
caas = caas.replace({ 'framework': 'NodeJs', 'rel_db': 'postgres', 'mem_db': 'Redis'})
caas
```

```python
# Rename the columns
iaas.columns = ['users', 'component', 'instances', 'RAM', 'CPU', 'STORAGE', 'Disk_type', 'technology']
caas.columns = ['users', 'component', 'instances', 'RAM', 'CPU', 'STORAGE', 'Disk_type', 'technology']
iaas_total.columns = ['users', 'iaas instances', 'storage_iaas']
caas_total.columns = ['use', 'caas instances', 'storage_caas']
```

```python
# Concatenate the dataframes
frames = [iaas, caas]
df = pd.concat(frames, ignore_index=True)
# df = pd.concat([df, disk_type_high], ignore_index=True, axis=1)
# df.columns = ['users', 'component', 'instances', 'RAM', 'CPU', 'STORAGE', 'Disk_type', 'technology']
df.to_csv(f'datasets/user_component_randm.csv')

totals = pd.concat([iaas_total, caas_total, disk_type], axis=1)
totals = totals.rename(columns={0: 'Disk_type'})
totals = totals.drop(columns='use')
totals.to_csv(f'datasets/total_instances.csv')

df
```

```python
# %% ================ Visualization ========================
#
# Grouped figured with averaged value
fig = px.histogram(df, x="technology", y=['RAM', 'CPU', 'STORAGE'] ,log_y=True,
                   barmode='group',  histfunc='avg')

fig.update_layout(
    title="Resources IaaS vs CaaS",
    xaxis_title="Cloud service design",
    yaxis_title="Average resources per users",
    legend_title="Resources",
    legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01
),
    font=dict(
        family="Courier New, monospace",
        size=32,
        color="RebeccaPurple"
    )
)

fig.show()



# fig.write_html("resources_compare.html")
# %%
# fig2 = px.scatter(df, x="users", y=['RAM', 'CPU', 'STORAGE'], facet_col='size', log_y=True)
# fig2.show()
# fig2.write_html("users_resources.html")
```
